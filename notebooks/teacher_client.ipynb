```python
import requests
from transformers import AutoTokenizer
from peft import PeftConfig

STUDENT_MODEL_NAME = "./knowledge-ingestion-test/model/checkpoint-178"
config = PeftConfig.from_pretrained(STUDENT_MODEL_NAME)

tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
inputs = tokenizer("Hello, how are you?")
inputs
```

```python
response = requests.post("http://localhost:8000/logprobs", json={"input_ids": [inputs["input_ids"]], "attention_mask": [inputs["attention_mask"]]})
response.json()
```
